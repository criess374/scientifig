```python
import numpy as np
import scientifig
import matplotlib.pyplot as plt
```

```python
scientifig.use_style("poster", width=1, background="transparent")
fig, ax = scientifig.create_figure(figsize=(20, 8))

# generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label="Sine Wave", color="blue")
ax.set_title("Sine Wave Plot")
# add xlabel
ax.set_xlabel("X-axis")
# add ylabel
ax.set_ylabel("Y-axis")
plt.legend()
# make annotation in red at x=5, y=0
ax.annotate(
    "Annotation",
    xy=(5, 0),
    xytext=(6, 0.5),
    arrowprops=dict(facecolor="red", shrink=0.05),
)
scientifig.scale_fonts(fig)
scientifig.savefig(fig, "testplots/sine_wave_plot")
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature

scientifig.use_style("paper", width=1.0)

fig, ax = scientifig.create_figure(
    figsize=(15, 5), subplot_kw={"projection": ccrs.PlateCarree()}
)

ax.add_feature(cfeature.LAND, facecolor="lightgray")
ax.add_feature(cfeature.OCEAN, facecolor="lightblue")
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linewidth=0.4, linestyle="--")
ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())
gl = ax.gridlines(draw_labels=True, color="gray", linestyle="--")
gl.top_labels = False
gl.right_labels = False
ax.set_title("World Map", y=1.05)

scientifig.scale_fonts(fig)
scientifig.savefig(fig, "testplots/world_map")
plt.show()
```
