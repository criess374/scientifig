```python
import numpy as np
import scientifig
import matplotlib.pyplot as plt
```

```python
# basic demo of using scientifig to create a plot with a custom style

scientifig.use_style("paper", width=0.5, background="black")
fig, ax = scientifig.create_figure()

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label="Sine Wave")
ax.set_title("Sine Wave Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")


# add legend at top right
plt.legend(loc="lower right")
# scientifig.reset_style()
# scientifig.savefig(fig, "testplots/sine_wave_plot")
plt.show()
```

```python
# demo over overwriting specific settings --> use with caution as it will limit the package's ability to automatically adjust the figure to fit the content

scientifig.use_style("paper", width=1, background="transparent")
fig, ax = scientifig.create_figure()

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label="Sine Wave", color="orange", linewidth=2)
ax.set_title("Sine Wave Plot", fontsize=32)
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

ax.annotate(
    "Annotation",
    xy=(5, 0),
    xytext=(6, 0.5),
    arrowprops=dict(shrink=0.05),
)
# add legend at top right
plt.legend(loc="lower right")
scientifig.savefig(fig, "testplots/sine_wave_plot.png")
plt.show()
```

```python
# demo shopwing resetting style

scientifig.reset_style()
fig, ax = plt.subplots()

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label="Sine Wave")
ax.set_title("Sine Wave Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")


# add legend at top right
plt.legend(loc="lower right")

plt.show()
```

```python
# demo of apply_style: retroactively style a figure that wasn't made with create_figure

fig, ax = plt.subplots()  # plain matplotlib figure, not scientifig.create_figure

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label="Sine Wave")
ax.set_title("Sine Wave Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend(loc="lower right")

# fonts/line width/background weren't set at creation time, so apply them now
# same theme/width/background as the first demo figure, so the two should
# render identically (create_figure defaults to matplotlib's own figsize too)
scientifig.use_style("paper", width=0.5, background="black")
scientifig.apply_style(fig)

plt.show()
```

```python
# demo using cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature

scientifig.use_style("paper", width=1.0, background="black")

fig, ax = scientifig.create_figure(
    figsize=(5, 5), subplot_kw={"projection": ccrs.PlateCarree()}
)

ax.add_feature(cfeature.LAND, facecolor="lightgray")
ax.add_feature(cfeature.OCEAN, facecolor="lightblue")
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linewidth=0.4, linestyle="--")
ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())
gl = ax.gridlines(draw_labels=True, color="gray", linestyle="--")
gl.top_labels = False
gl.right_labels = False
ax.set_title("World Map", y=1.05)


# scientifig.savefig(fig, "testplots/world_map")
plt.show()
```

```python
# demo of using a custom style

from pathlib import Path

scientifig.THEMES["crazy"] = {
    "mplstyle": Path(
        "C:/Users/riec/Documents/scientifig/scientifig/mplstyle/crazy_style.mplstyle"
    ),
    "canvas_width_in": 6.0,
}
scientifig.use_style("crazy", width=1.0)


fig, ax = scientifig.create_figure()

x = np.linspace(0, 10, 100)

for i in range(6):
    y = np.sin(x * (1 + 0.15 * i) + i) + 0.3 * i
    ax.plot(x, y, label=f"Sine {i + 1}")

ax.set_title("Sine Wave crazy Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.legend()


scientifig.savefig(fig, "testplots/sine_wave_plot.png")
plt.show()
```
