```python
import numpy as np
import scientifig
import matplotlib.pyplot as plt
```

```python
scientifig.use_style(
    "paper", width=1, background="black"
)  # once, e.g. at the top of a notebook

fig, ax = scientifig.create_figure(figsize=(6, 4))

# generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label="Sine Wave", color="blue")
ax.set_title("Sine Wave Plot")
plt.legend()
scientifig.scale_fonts(fig)
# scientifig.savefig(fig, "testplots/sine_wave_plot.png")
```
