```python
import numpy as np
import matplotlib.pyplot as plt
import scientifig
```

```python
scientifig.use_style(
    "presentation", width=0.2, background="black"
)  # once, e.g. at the top of a notebook

fig, ax = scientifig.create_figure(figsize=(7.5, 6))

# generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label="Sine Wave", color="blue")

scientifig.scale_fonts(fig)
plt.show()
```
